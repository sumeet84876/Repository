Question 3: What is clustering in unsupervised learning? Mention three popular clustering algorithms.

Answer: Clustering is an unsupervised learning technique used to group similar data points into clusters without using any labels.
The goal is to ensure that data points in the same cluster are more similar to each other than to those in other clusters.

Three popular clustering algorithms:

K-Means Clustering – Divides data into k predefined clusters based on distance.

Hierarchical Clustering – Builds a tree-like structure (dendrogram) to form clusters.

DBSCAN (Density-Based Spatial Clustering of Applications with Noise) – Groups points based on density and detects noise/outliers.

Question 8: Write Python code to perform K-means clustering on a sample dataset.

```python
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
X, _ = make_blobs(n_samples=200, centers=3, random_state=42)
kmeans = KMeans(n_clusters=3)
kmeans.fit(X)
print(kmeans.labels_)
```

Question 9: What is inheritance in OOP? Provide a simple example in Python?

Answer: Inheritance in Object-Oriented Programming (OOP) is a feature that allows one class (child class) to acquire properties and methods of another class (parent class).
It helps in reusability, clean code, and avoiding duplication.

```python
# Parent class
class Animal:
    def sound(self):
        return "Animal makes a sound"

# Child class inheriting from Animal
class Dog(Animal):
    def sound(self):
        return "Dog barks"

# Object of child class
d = Dog()
print(d.sound())
```
