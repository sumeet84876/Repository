4. A How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap?

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = {
    "A": [1, 2, 3, 4, 5],
    "B": [5, 4, 3, 2, 1],
    "C": [2, 3, 4, 5, 6]
}
df = pd.DataFrame(data)
corr = df.corr()
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix Heatmap")
plt.show()
```

8. Create a line plot with multiple lines using Matplotlib?

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y1 = [2, 4, 6, 8, 10]
y2 = [1, 3, 5, 7, 9]
y3 = [2, 3, 4, 5, 6]

plt.plot(x, y1, label="Line 1", color="blue", marker="o")
plt.plot(x, y2, label="Line 2", color="green", marker="s")
plt.plot(x, y3, label="Line 3", color="red", linestyle="--")

plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Line Plot with Multiple Lines")
plt.legend()

plt.show()
```

13. Create a 3D scatter plot using Plotly?

```python
import plotly.express as px
import pandas as pd

data = {
    "x": [1, 2, 3, 4, 5],
    "y": [10, 11, 12, 13, 14],
    "z": [5, 6, 7, 8, 9],
    "category": ["A", "A", "B", "B", "C"]
}

df = pd.DataFrame(data)

fig = px.scatter_3d(df, x="x", y="y", z="z", color="category", size=[10,20,30,40,50])

fig.show()
```
