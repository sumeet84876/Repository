Q1: What is hypothesis testing in statistics?
Answer: Hypothesis testing is a statistical method to make decisions about a population parameter using sample data. It tests whether there is enough evidence to reject a stated assumption (null hypothesis).

Q2: What is the null hypothesis, and how does it differ from the alternative hypothesis?
Answer:

Null Hypothesis (H₀): Assumes no effect or no difference (status quo).

Alternative Hypothesis (H₁): Assumes there is an effect or difference.
The test checks if data gives enough evidence to reject H₀ in favor of H₁.


Q3: Explain the significance level in hypothesis testing and its role in deciding the outcome of a test.
Answer: The significance level (α, usually 0.05) is the probability of rejecting the null hypothesis when it is actually true (Type I error). If the p-value ≤ α, we reject H₀; otherwise, we fail to reject H₀.

Q4: What are Type I and Type II errors? Give examples of each.
Answer:

Type I Error (False Positive): Rejecting the null hypothesis when it is actually true.
Example: Concluding a medicine works when it actually doesn’t.

Type II Error (False Negative): Failing to reject the null hypothesis when it is actually false.
Example: Concluding a medicine doesn’t work when it actually does.


Q5: What is the difference between a Z-test and a T-test? Explain when to use each.
Answer:

Z-test: Used when population variance (σ²) is known and sample size is large (n > 30).

T-test: Used when population variance is unknown and sample size is small (n ≤ 30)



Q6: Write a Python program to generate a binomial distribution with n=10 and
p=0.5, then plot its histogram.

```python
import numpy as np
import matplotlib.pyplot as plt


n = 10
p = 0.5

data = np.random.binomial(n=n, p=p, size=1000)

plt.hist(data, bins=range(n+2), edgecolor='black', align='left')
plt.title("Binomial Distribution (n=10, p=0.5)")
plt.xlabel("Number of Successes")
plt.ylabel("Frequency")
plt.show()
```

Q7: Implement hypothesis testing using Z-statistics for a sample dataset in
Python. Show the Python code and interpret the results.
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

```python
import numpy as np
import scipy.stats as st

sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

data = np.array(sample_data)

mu = 50
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)


n = len(data)


z_stat = (sample_mean - mu) / (sample_std / np.sqrt(n))


p_value = 2 * (1 - st.norm.cdf(abs(z_stat)))

print("Sample Mean:", sample_mean)
print("Z-Statistic:", z_stat)
print("P-value:", p_value)


alpha = 0.05
if p_value < alpha:
    print("Reject Null Hypothesis (H0).")
else:
    print("Fail to Reject Null Hypothesis (H0).")
```

Q8: Write a Python script to simulate data from a normal distribution and
calculate the 95% confidence interval for its mean. Plot the data using Matplotlib?

=>We simulate random data from a normal distribution, then compute the sample mean and its 95% confidence interval using the t-distribution. Finally, we plot the histogram to visualize the data.

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as st


data = np.random.normal(loc=100, scale=15, size=500)

mean_val = np.mean(data)

sem = st.sem(data)

confidence_interval = st.t.interval(0.95, len(data)-1, loc=mean_val, scale=sem)

print("Sample Mean:", mean_val)
print("95% Confidence Interval:", confidence_interval)

plt.hist(data, bins=30, edgecolor='black', alpha=0.7)
plt.axvline(mean_val, color='red', linestyle='dashed', linewidth=2, label=f"Mean = {mean_val:.2f}")
plt.axvline(confidence_interval[0], color='green', linestyle='dashed', linewidth=2, label="95% CI Lower")
plt.axvline(confidence_interval[1], color='blue', linestyle='dashed', linewidth=2, label="95% CI Upper")
plt.title("Normal Distribution with 95% Confidence Interval")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```
